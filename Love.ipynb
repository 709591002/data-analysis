```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X=np.arange(-5.0,5.0,0.1)
```

```python
Y=np.arange(-5.0,5.0,0.1)
```

```python
x,y=np.meshgrid(X,Y)
f=17*x**2-16*np.abs(x)*y+17*y**2-225
```

```python
fig=plt.figure()
cs=plt.contour(x,y,f,0,colors='r')
plt.show()
```
